## TASK 3: Relationship between Film Revenue and Budget

```python
sns.scatterplot(data['revenue'],data['budget'])
plt.title("Relationship b/w Film Revenue and Film Budget")
plt.show()
```

```python
data.plot(x='budget',y='revenue',style='.',color='palevioletred')
plt.title('Relationship b/w Film Revenue and Film Budget')
plt.xlabel('BUDGET')
plt.ylabel('REVENUE')
plt.show()
```

```python
plt.figure(figsize=(10,10))
sns.heatmap(data.corr())
plt.show()
```

```python
relation=data[['revenue','budget']].corr
sns.heatmap(relation())
plt.show()
```

## TASK 4: Distribution of Languages in Films

```python
data['original_language'].unique()
sns.scatterplot(data['original_language'],data['revenue'],color='darkkhaki')
plt.show()
```

```python
sns.scatterplot(data['original_language'],np.log1p(data['revenue']),color='peru')
plt.show()
```

```python
sns.boxplot(data['original_language'],data['revenue'])
plt.show()
```

```python
data[data['original_language'] == data['original_language'].max()]
```

```python
plt.subplots(figsize=(10,10))
data.groupby(['original_language'])['revenue'].sum().sort_values(ascending = False).head(20).plot(kind = 'pie')
plt.title("Distribution of languages in Films", fontsize=24)
plt.xlabel('original language', fontsize=18)
plt.ylabel('revenue', fontsize=18)
plt.xticks(rotation = 90)
plt.show()
```
